```python
import json

data_list = []
with open('../data/daccord_yifei_v2_leo_supp_anti_oth.json', 'r') as input_file:
    for line in input_file:
        data = json.loads(line)
        if data["User label"] == 'others':
            data["User label"] = 'other'
        if data["User label"] == 'anti':
            data["User label"] = 'anti gun'
        if data["User label"] == 'support':
            data["User label"] = 'support gun'
        
        data_list.append(data)
```

```python
data_list
```

```python
with open('../data/daccord_yifei_v2_leo_supp_anti_oth.json', 'w') as output_file:
    for data_dict in data_list:
        # 将每个字典写入JSON文件，每行一个JSON对象
        json.dump(data_dict, output_file)
        output_file.write('\n')  # 写入换行符，以便下一个对象位于新行
```
